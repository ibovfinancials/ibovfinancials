```python
# Importa as bibliotecas `requests` para fazer requisições HTTP e `pandas` para manipular dados em DataFrames

import requests
import pandas as pd

# Define o símbolo da ação que queremos consultar (neste caso, "azul4")
symbol = 'azul4'

# Substitua 'your token here' pelo seu token real da API do Ibovespa Financials (já está preenchido aqui)
token = 'your token here'

# Define a URL base da API do Ibovfinancials
url = 'https://ibovfinancials.com'

# Define o endpoint específico para consulta de dados históricos
endpoint = '/api/ibov/historical'

# Cria um dicionário de parâmetros para a requisição
params = {
  "symbol": symbol,  # Símbolo da ação a ser consultada
  "timeframe": 5,   # Timeframe dos dados (5 = diário, 60 = mensal, 120 = anual)
  "start_date": "2024-08-01",  # Data inicial no formato 'YYYY-MM-DD'
  "end_date": "2024-08-15",   # Data final no formato 'YYYY-MM-DD'
  "token": token,    # Token de acesso à API
}

# Realiza a requisição GET para o endpoint da API
response = requests.get(url + endpoint, params=params)

# Converte a resposta da API para um objeto JSON
response_json = response.json()
```

```python
data = pd.DataFrame(response_json['AZUL4'])
data
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=data['time'],
                open=data['open'],
                high=data['high'],
                low=data['low'],
                close=data['close'])])

fig.update_layout(title='Candle Plot', xaxis_title='Time', yaxis_title='Price')

fig.show()
```
